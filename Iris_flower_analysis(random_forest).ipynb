## IRIS FLOWER ANALYSIS (Random Forest Classifier)

```python
#loading the dataset
from sklearn.datasets import load_iris
#loading the random forest classifier
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
#setting random seed
np.random.seed(0)
```

```python
#creating object for the load_iris
iris=load_iris()
```

```python
#loading to the pandas dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)
#view first 5 rows
df.head()
```

```python
#Adding a new column for the species name
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
df.head()
```

# Creating Train and Test Data

```python
df['is_train']=np.random.uniform(0,1,len(df))<=.75
df.head()
```

```python
#creating dataframes with test rows and training rows
train, test=df[df['is_train']==True], df[df['is_train']==False]
train.shape,test.shape
```

```python
#show the no.of observations
print('Number of observation in the training data :',len(train))
print('Number of observation in the test data :',len(test))
```

```python
#create a list of the feature columes names
features=df.columns[:4]
#view features
features
```

## converting each species name into digits

```python
y=pd.factorize(train['species'])[0]
y
```

## creating a Random Forest Classifier

```python
# Create a random forest Classifier. By convention, clf means 'Classifier'
clf = RandomForestClassifier(n_jobs=2, random_state=0)

# Train the Classifier to take the training features and learn how they relate
# to the training y (the species)
clf.fit(train[features], y)
```

```python
#Applying the trained classifier to the test
clf.predict(test[features])
```

```python
#viewing the predicted probabilities of first 10 observations
clf.predict_proba(test[features])[0:10]
```

```python
#mapping names for the plants for each predicted plant class
preds=iris.target_names[clf.predict(test[features])]
#view the predicted species
preds[0:5]
```

```python
#viewing the actual species
test['species'].head()
```

## creating confusion matrix

```python
pd.crosstab(test['species'],preds,rownames=['Actual species'],colnames=['Predicted species'])
```
